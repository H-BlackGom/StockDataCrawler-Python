### DataReader에서 NaN 데이터 발생에 따른 라이브러리 변경 테스트

```python
import yfinance as yf
from tqdm.notebook import tqdm
from datetime import datetime

# now_date = datetime.now().strftime('%Y-%m-%d')
save_stock_df = pd.DataFrame()
for idx, (code, company) in tqdm(enumerate(company_codes), total=len(company_codes)):
    msft = yf.Ticker(code)
    stock_data = msft.history(period="1d")
    stock_data.reset_index(inplace=True)
    stock_data['Company'] = company
    stock_data['Code'] = code
    stock_data['Type'] = stock_data.apply(lambda x: int(x['Close'] > x['Open']), axis=1)
    stock_data['candleCenter'] = (stock_data.Open + stock_data.Close) / 2
    
    save_stock_df = pd.concat([save_stock_df, stock_data])

save_stock_df.drop(["Dividends"], axis=1, inplace=True)
save_stock_df.drop(["Stock Splits"], axis=1, inplace=True)
save_stock_df
```

```python
conn = MongoClient('192.168.0.12', 27017)
collection_stock = conn['Changsung']['stock_test']
collection_stock.insert_many(save_stock_df.to_dict('records'))
```
